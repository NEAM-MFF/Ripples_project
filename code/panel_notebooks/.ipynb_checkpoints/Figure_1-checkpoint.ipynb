```python
def calculate_ripple_rates_fig1(monkey_list,dual_th=[2.5,3.5],area=None,params={},
                              main_folder='/CSNG/studekat/ripple_band_project/',df_folder='',verbose=False):
    """
    Calculates ripple rates for each channel, pooling all monkeys in the list, all dates.
    Area can be chosen: V1, V2, V12, V4, IT.

    Returns a dict with lists of rates for channels in all Rec, EC, and EO.
    """
    condition = 'RS'
    
    ripple_rates_ch_list = []
    EC_ripple_rates_ch_list = []
    EO_ripple_rates_ch_list = []
    for monkey in monkey_list:
        all_dates = params['dates'][monkey][condition]
        for date in all_dates:
            df_r = load_ripples_df(monkey,dual_th=dual_th,date=date,area=area,condition=condition,
                                   params=params_analysis,df_folder=df_folder,exclude_noisy=True,verbose=verbose)  # only including good channels
            ### Duration of the whole recording, and EC and EO periods
            times = params['times_all_arr'][monkey][condition][date]
            dur_rec_date_ms = times[1] - times[0]
            dur_rec_date_s = dur_rec_date_ms/1000
            file_path = f'{main_folder}/metadata/EC_EO_indicators/eyes_indic_monkey_{monkey}_RS_date_{date}_common_times.pkl'
            with open(file_path, 'rb') as file:
                eyes_dict = pickle.load(file)
            EC_dur_s = np.sum(eyes_dict['EC'])/1000
            EO_dur_s = np.sum(eyes_dict['EO'])/1000
            if area is not None:
                if area in ['V4','IT','V1','V2']:
                    all_arrays = np.where(np.array(params['areas'][monkey])==area)[0] + 1  # shifting indexing from 0 to 1
                if area in ['V12']:
                    all_arrays = np.where(np.isin(np.array(params['areas'][monkey]),['V1','V2']))[0] + 1 
            else:
                all_arrays = range(1,17)
            for array in all_arrays:
                df_array = df_r[df_r['array']==array]
                if df_array.shape[0]>0:
                    for ch in np.unique(df_array['channel_0_63']): 
                        df_ch = df_array[df_array['channel_0_63']==ch]
                        num_ripples_ch = df_ch.shape[0]
                        df_ch = df_ch[df_ch['eyes_closed'].notna()]  # not using non-classified times of recording
                        df_ch_EC = df_ch[df_ch['eyes_closed']]
                        num_ripples_ch_EC = df_ch_EC.shape[0]
                        df_ch_EO = df_ch[np.logical_not(df_ch['eyes_closed'])]
                        num_ripples_ch_EO = df_ch_EO.shape[0]
                        ripple_rates_ch_list.append(num_ripples_ch/dur_rec_date_s)
                        EC_ripple_rates_ch_list.append(num_ripples_ch_EC/EC_dur_s)
                        EO_ripple_rates_ch_list.append(num_ripples_ch_EO/EO_dur_s)
                else:
                    if verbose:
                        print(f'No data for monkey {monkey}, array {array}.')

    rates_dict = {'All': ripple_rates_ch_list,
                 'EC': EC_ripple_rates_ch_list,
                 'EO': EO_ripple_rates_ch_list}
    return rates_dict
```

```python
def calculate_ripple_prop_fig1(monkey_list,dual_th = [2.5,3.5],prop='duration_ms',area=None,params={},
                              main_folder='/CSNG/studekat/ripple_band_project/',df_folder='',verbose=False):
    """
    Calculates ripple prop. as chosen, pooling all monkeys in the list, all dates.
    Area can be chosen.

    prop: duration_ms, amplitude, freq, CV ICI

    Returns a dict with lists of the values of a given prop. for channels in all Rec, EC, and EO.
    """
    condition = 'RS'
    
    ripple_prop_list = []
    EC_ripple_prop_list = []
    EO_ripple_prop_list = []
    for monkey in monkey_list:
        all_dates = params['dates'][monkey][condition]
        for date in all_dates:
            df_r = load_ripples_df(monkey,dual_th=dual_th,date=date,area=area,condition=condition,
                                   params=params_analysis,df_folder=df_folder,exclude_noisy=True)  # only including good channels
            if area is not None:
                if area in ['V4','IT','V1','V2']:
                    all_arrays = np.where(np.array(params['areas'][monkey])==area)[0] + 1  # shifting indexing from 0 to 1
                if area in ['V12']:
                    all_arrays = np.where(np.isin(np.array(params['areas'][monkey]),['V1','V2']))[0] + 1 
            else:
                all_arrays = range(1,17)
            for array in all_arrays:
                df_array = df_r[df_r['array']==array]
                if df_array.shape[0]>0:
                    ripple_prop_list.append(df_array[prop].values)
                    df_array = df_array[df_array['eyes_closed'].notna()]  # not using non-classified times of recording
                    df_array_EC = df_array[df_array['eyes_closed']]
                    df_array_EO = df_array[np.logical_not(df_array['eyes_closed'])]
                    EC_ripple_prop_list.append(df_array_EC[prop].values)
                    EO_ripple_prop_list.append(df_array_EO[prop].values)
                else:
                    if verbose:
                        print(f'No data for monkey {monkey}, array {array}.')

    prop_dict = {'All': list_merge(ripple_prop_list),
                 'EC': list_merge(EC_ripple_prop_list),
                 'EO': list_merge(EO_ripple_prop_list)}
    return prop_dict
```

```python
def plot_ripple_compare_EC_EO(dict_animals,set_params):
    """
    PANEL 1

    Figure comparing properties between animals + pool.
    """
    fig, ax = plt.subplots()
    fig.set_figwidth(set_params['plot_width'])
    fig.set_figheight(set_params['plot_height'])

    plt.suptitle(set_params['plot_suptitle'])
    
    # Pool plot
    data_list = []
    for monkey, subdict in dict_animals.items():
        for option, values in subdict.items():  # all, EC, EO 
            for val in values:
                data_list.append({
                    'Option': option,
                    'Data': val
                })
        df_long = pd.DataFrame(data_list)
        
    sns.violinplot(ax=ax,data=df_long, x='Option', y='Data',hue='Option',
                   palette=set_params['eyes_colors'],cut=0,bw_method=set_params['violin_bw'])    
    for violin in ax.collections: 
        violin.set_alpha(set_params['violin_alpha'])

    ax.set_ylim(set_params['ylim'])
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.set_ylabel(set_params['x_label'])
    ax.set_xlabel('')

    pairs = [('EC', 'EO')]
    annotator = Annotator(ax, pairs, data=df_long, x='Option', y='Data')
    annotator.configure(
        test='Mann-Whitney',
        text_format='star',
        comparisons_correction=None,
        loc='outside',
        verbose=False,
    )
    annotator.apply_and_annotate()
    
    plt.tight_layout()
    plt.show()
    plt.close()
    return
```

```python
def plot_ripple_start_trigg_EC_EO(df_ripple_start_trig,type_sig='LFP',set_params={}):
    """
    PANEL 1
    
    Average ripple triggered LFP/RB (LFP, RB_sig type_sig).
    """
    fig, ax = plt.subplots(1,3,sharex=True,sharey=True)
    fig.set_figwidth(set_params['plot_width'])
    fig.set_figheight(set_params['plot_height'])

    plt.suptitle(f'RB trig. LFP')

    options = ['_EC','_EO','']
    labels = {'_EC': 'EC',
              '_EO': 'EO',
              '': 'All',
    }
    for idx in range(3):
        option = options[idx]
        column_name = f'{type_sig}_trigg_mean{option}'
        print(column_name)
        valid_rows = df_ripple_start_trig[column_name].apply(lambda x: isinstance(x, np.ndarray) and x.shape==(20000,))
        filtered_values = df_ripple_start_trig.loc[valid_rows, column_name].to_numpy()
    
        RB_sig_mean = np.nanmean(np.vstack(filtered_values),axis=0)
        buffer = set_params['buffer']
        if option in ['_EC','_EO']:
            color_option = set_params[f'colors_ripple_classes{option}']
        else:
            color_option = 'gray'
        
        x = np.arange(-buffer,buffer)
        
        ax[idx].plot(x,RB_sig_mean[10000-buffer:10000+buffer],color=color_option,label=labels[option],linewidth=2)
        ax[idx].spines['top'].set_visible(False)
        ax[idx].spines['right'].set_visible(False)
        ax[idx].set_xlabel('Time [ms]')
        if idx==0:
            ax[idx].set_ylabel(f'Avg. {type_sig} [norm. u.]')
        ax[idx].axvline(0,ls='dashed',color='gray',alpha=0.7)
        #ax[idx].legend()
    fig.legend(loc='upper right', bbox_to_anchor=(1.1, 1))

    plt.tight_layout()
    plt.show()
    plt.close()
    return 
```

```python
def plot_ripple_compare_EC_EO_animals(dict_animals,set_params):
    """
    PANEL 1

    Figure comparing properties between animals, EC and EO.
    """
    fig, ax = plt.subplots()
    fig.set_figwidth(set_params['plot_width'])
    fig.set_figheight(set_params['plot_height'])

    plt.suptitle(set_params['plot_suptitle'])

    for monkey in dict_animals.keys():
        #print(monkey)
        medians = []
        for pos, option in enumerate(['All','EC','EO']):
            df = pd.DataFrame(dict_animals[monkey][option])
            df_long = df.melt(var_name=option, value_name='Value')
            
            boxx = sns.boxplot(x=option, y='Value',positions=[pos], data=df_long, color=set_params['colors_monkeys'][monkey],showfliers=False,
                                 boxprops=dict(facecolor="none", edgecolor=set_params['colors_monkeys'][monkey], linewidth=1.5),whis=0,
                                 medianprops=dict(color=set_params['colors_monkeys'][monkey], linewidth=1.5),width=0.5,
                                whiskerprops=dict(visible=False),
                                capprops=dict(visible=False))
            medians.append(df_long['Value'].median())

        for line in boxx.lines:
            line.set_alpha(0.7)
        for patch in boxx.patches:
            patch.set_alpha(0.7)
        
        x_positions = range(len(medians))
        plt.plot(x_positions, medians, color=set_params['colors_monkeys'][monkey], marker='o', linestyle='-',label=monkey)

    ax.set_ylim(set_params['ylim'])
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.set_ylabel(set_params['x_label'])
    ax.set_xlabel('')
    ax.set_xticks([0,1,2])
    ax.set_xticklabels(['All','EC','EO'])

    plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
    plt.tight_layout()
    plt.show()
    plt.close()
    return
```

```python
def plot_ch_rates_per_animal(rates_ratio_EC_EO_dict,monkey_colors=COLORS_MONKEYS):
    """
    Plotting the distribution of EC or EO pref. channels in ripples per animal.
    """
    fig, ax = plt.subplots()
    fig.set_figwidth(2)
    fig.set_figheight(2)

    for monkey in rates_ratio_EC_EO_dict.keys():
        #print(monkey)
        data = rates_ratio_EC_EO_dict[monkey]
        data = data[np.isfinite(data)]

        sns.kdeplot(data, bw_adjust=1, fill=True, color=monkey_colors[monkey],label=monkey)
        
    ax.axvline(1,ls=':',alpha=0.8,color='k',label='EC=EO')
    #ax.axvline(np.median(data),ls='dashed',alpha=0.8,color='darkblue',label='median')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.set_xlabel('EC/EO ripple ratio per ch. \n[norm. u.]')
    ax.set_ylabel('Density [norm. u.]')
    #ax.set_xscale('log')
    #ax.set_xlim([0.1,5])
    
    #ax.text(0.90, 0.98,"More EC",ha='right', va='top', fontsize=18, color='darkblue',transform=ax.transAxes)
    
    plt.legend()
    plt.show()
    plt.close()
    return
```

### Data loading

```python
lowpass_freq = 40
MAIN_FOLDER = '/CSNG/studekat/ripple_paper_clean/'
```

```python
### RIPPLE ENVELOPE LOADING ###

# For each monkey, we have three dictionaries: All, EC, EO. In each of them, dictionary with array-wise ripple envelope avg. np array
# CAREFUL - WE NEED TO CUT OUT THE COMMON TIMES

dates_list = list_merge([DATES['L']['RS'],DATES['N']['RS'],DATES['F']['RS']])
monkeys_list = ['L','L','L','N','N','F','F','F']
all_env_dict = {k:{} for k in dates_list}

for date, monkey in zip(all_env_dict.keys(), monkeys_list):
    print(date, monkey)
    all_env_dict[date] = {k: {} for k in ['All','EC','EO']}
    for array in range(1,17):
        if AREAS[monkey][array-1] in ['V1','V2']:
            try:
                # loading RB block (+ 40 Hz lowpass) and cutting into common times, EC and EO
                file_path = f'{MAIN_FOLDER}/metadata/EC_EO_indicators/eyes_indic_monkey_{monkey}_RS_date_{date}_common_times.pkl'
                with open(file_path, 'rb') as file:
                    eyes_dict = pickle.load(file)
                
                RB_block = load_block(monkey,array,type_rec='RS',type_sig='RB',date=date,data_folder=DATA_FOLDER)
                RB_sig_array = sig_block_to_arr(RB_block,'RB_filtered_zsc')
                RB_env_array = sig_block_to_arr(RB_block,'RB_envelope_norm')
            
                if lowpass_freq is not None:
                    RB_env_array = elephant.signal_processing.butter(RB_env_array, highpass_frequency=None, lowpass_frequency=lowpass_freq, 
                                                                order=6, filter_function='sosfiltfilt',sampling_frequency=1000, axis=-1)
                    # renormalize, because we filtered again
                    RB_env_array = RB_env_array/RB_env_array.std(axis=1,keepdims=True)
                
                n_channels = 64
                start_rec = float(RB_block.segments[0].analogsignals[0].t_start.rescale('ms').magnitude)
                stop_rec = float(RB_block.segments[0].analogsignals[0].t_stop.rescale('ms').magnitude)
                #print(start_rec)
    
                EC_indicator = eyes_dict['EC'].astype(bool)
                EO_indicator = eyes_dict['EO'].astype(bool)
    
                RB_env_array = cut_abs_times(RB_env_array,np.int64(start_rec),monkey,rec_type='RS',date=date,params=params_analysis)
                all_env_dict[date]['All'][array] = np.sum(RB_env_array,axis=0)
                all_env_dict[date]['EC'][array] = np.sum(RB_env_array,axis=0)[EC_indicator]
                all_env_dict[date]['EO'][array] = np.sum(RB_env_array,axis=0)[EO_indicator]
                print(array)
            except:
                print(f'The data for array {array} could not be loaded.')
        else:
            print(f'Array {array} is not in V12.')
        
```

```python
### DELTA FILTERED LOADING ###

# For each monkey, we have three dictionaries: All, EC, EO. In each of them, dictionary with array-wise ripple envelope avg. np array
# CAREFUL - WE NEED TO CUT OUT THE COMMON TIMES

dates_list = list_merge([DATES['L']['RS'],DATES['N']['RS'],DATES['F']['RS']])
monkeys_list = ['L','L','L','N','N','F','F','F']
all_delta_dict = {k:{} for k in dates_list}

for date, monkey in zip(all_env_dict.keys(), monkeys_list):
    print(date, monkey)
    all_delta_dict[date] = {k: {} for k in ['All','EC','EO']}
    for array in range(1,17):
        if AREAS[monkey][array-1] in ['V1','V2']:
            try: 
                # loading LFP block (+ 40 Hz lowpass) and cutting into common times, EC and EO
                file_path = f'{MAIN_FOLDER}/metadata/EC_EO_indicators/eyes_indic_monkey_{monkey}_RS_date_{date}_common_times.pkl'
                with open(file_path, 'rb') as file:
                    eyes_dict = pickle.load(file)
                
                LFP_block = load_block(monkey,array,type_rec='RS',type_sig='LFP',date=date,data_folder=DATA_FOLDER)
                LFP_sig_array = sig_block_to_arr(LFP_block,'LFP_zsc')
            
                delta_array = elephant.signal_processing.butter(LFP_sig_array, highpass_frequency=0, lowpass_frequency=4, 
                                                            order=6, filter_function='sosfiltfilt',sampling_frequency=1000, axis=-1)

                n_channels = 64
                start_rec = float(LFP_block.segments[0].analogsignals[0].t_start.rescale('ms').magnitude)
                stop_rec = float(LFP_block.segments[0].analogsignals[0].t_stop.rescale('ms').magnitude)
                #print(start_rec)
    
                EC_indicator = eyes_dict['EC'].astype(bool)
                EO_indicator = eyes_dict['EO'].astype(bool)
    
                delta_array = cut_abs_times(delta_array,np.int64(start_rec),monkey,rec_type='RS',date=date,params=params_analysis)
                all_delta_dict[date]['All'][array] = np.sum(delta_array,axis=0)
                all_delta_dict[date]['EC'][array] = np.sum(delta_array,axis=0)[EC_indicator]
                all_delta_dict[date]['EO'][array] = np.sum(delta_array,axis=0)[EO_indicator]
                print(array)
            except:
                print(f'The data for array {array} could not be loaded.')
        else:
            print(f'Array {array} is not in V12.')
        
```

### Example of transition of dynamical regimes - use the ripple band envelope

```python
WIN_WIDTH = 15*1000
SIGMA = 15
monkey = 'L'

for date in list(all_env_dict.keys())[:1]:
    file_path = f'{MAIN_FOLDER}/metadata/EC_EO_indicators/eyes_indic_monkey_{monkey}_RS_date_{date}_common_times.pkl'
    with open(file_path, 'rb') as file:
        eyes_dict = pickle.load(file)

    EC_indic = eyes_dict['EC']
    arrays = list(all_env_dict[date]['All'].keys())
    num_wins = int(all_env_dict[date]['All'][arrays[0]].shape[0]//WIN_WIDTH)

    # total RB activity on all arrays
    all_rb = [all_env_dict[date]['All'][a] for a in arrays]
    all_rb = np.sum(np.vstack(all_rb),axis=0)
    all_rb = gaussian_filter1d(all_rb,sigma=SIGMA)
    #all_rb = (all_rb - all_rb.min())/(all_rb.max() - all_rb.min())
    
    # total Delta activity on all arrays
    all_delta = [all_delta_dict[date]['All'][a] for a in arrays]
    all_delta = np.sum(np.vstack(all_delta),axis=0)
    all_delta = gaussian_filter1d(all_delta,sigma=SIGMA)
    #all_delta = (all_delta - all_delta.min())/(all_delta.max() - all_delta.min())
   
    for i_win in range(num_wins):
        start_win = i_win*WIN_WIDTH
        stop_win = (i_win+1)*WIN_WIDTH
        
        fig, ax = plt.subplots()
        fig.set_figwidth(14)
        fig.set_figheight(6)

        plt.plot()

        x_pos = range(WIN_WIDTH)
        # TOTAL RB ACTIVITY and DELTA?
        all_delta_cut = all_delta[start_win:stop_win]
        all_rb_cut = all_rb[start_win:stop_win]
        all_delta_cut = (all_delta_cut - all_delta_cut.min())/(all_delta_cut.max() - all_delta_cut.min())
        all_rb_cut = (all_rb_cut - all_rb_cut.min())/(all_rb_cut.max() - all_rb_cut.min())
        
        ax.plot(x_pos,all_rb_cut,alpha=0.7,color='k',label='Total RB env.')
        ax.plot(x_pos,all_delta_cut,alpha=0.7,color='darkorange',label='Total Delta')
        
        # INIDIVIDUAL ARRAYS ACTIVITY
        shift = -1
        norm_factor = 2

        cmap = plt.cm.plasma
        colors = cmap(np.linspace(0, 0.8, len(arrays)))

        for idx, arr in enumerate(arrays):
            # loading
            arr_rb = all_env_dict[date]['All'][arr]
            arr_delta = all_delta_dict[date]['All'][arr]
            # smoothing
            arr_rb = gaussian_filter1d(arr_rb,sigma=SIGMA)
            arr_delta = gaussian_filter1d(arr_delta,sigma=SIGMA)
            # normalising
            arr_rb_min_max = (arr_rb - arr_rb.min()) / (arr_rb.max() - arr_rb.min())
            arr_delta_min_max = (arr_delta - arr_delta.min()) / (arr_delta.max() - arr_delta.min())
            # plotting
            ax.fill_between(x_pos,shift,arr_rb_min_max[start_win:stop_win]*norm_factor+shift,alpha=0.3,color=colors[idx])
            ax.plot(x_pos,arr_rb_min_max[start_win:stop_win]*norm_factor+shift,alpha=0.9,color=colors[idx])
            shift-=0.8
        
        for sp in ['top','right','left']:
            ax.spines[sp].set_visible(False)
        ax.set_yticks([])
        
        xticks_ms = ax.get_xticks()
        xticks_s = xticks_ms/1000
        ax.set_xticks(xticks_ms)
        ax.set_xticklabels([f'{x:.1f}' for x in xticks_s])

        ax.spines['right'].set_visible(False)
        ax.spines['top'].set_visible(False)

        plt.legend()
        plt.show()
        plt.close()

        fig, ax = plt.subplots()
        fig.set_figwidth(14)
        fig.set_figheight(1)

        plt.plot(EC_indic[start_win:stop_win])

        plt.show()
        plt.close()
```

### Delta trough triggered RB envelope

```python
CUT_WIN = 2000
EC_trigg_list = []
trough_list = []
delta_around_trigg = []
for date in all_delta_dict.keys():
    for arr in all_delta_dict[date]['All'].keys():
        EC_delta_vec = all_delta_dict[date]['EC'][arr]
        EC_env_vec = all_env_dict[date]['EC'][arr]
        len_vec = len(EC_env_vec)
        delta_troughs, _ = find_peaks(-EC_delta_vec)  # finding troughs, not peaks
        cut_around_troughs = [EC_env_vec[trough-CUT_WIN:trough+CUT_WIN] for trough in delta_troughs if (trough>CUT_WIN and trough<len_vec-CUT_WIN)]
        cut_delta_sig = [EC_delta_vec[trough-CUT_WIN:trough+CUT_WIN] for trough in delta_troughs if (trough>CUT_WIN and trough<len_vec-CUT_WIN)]
        EC_trigg_list.append(cut_around_troughs)
        trough_list.append(delta_troughs)
        delta_around_trigg.append(cut_delta_sig)

EC_trigg_list = list_merge(EC_trigg_list)
delta_trigg_list = list_merge(delta_around_trigg)
```

```python
trigg_env = np.sum(np.vstack(EC_trigg_list),axis=0)
trigg_delta = np.sum(np.vstack(delta_trigg_list),axis=0)
```

```python
fig, ax = plt.subplots()
fig.set_figheight(2)
fig.set_figwidth(2)

delta_vec = zscore(trigg_delta)[1000:3000]
env_vec = zscore(trigg_env)[1000:3000]

ax.plot(delta_vec,color='darkorange',label='Delta',linewidth=1.5)
ax.plot(env_vec,color='k',label='RB Envelope',linewidth=1.5)
ax.axvline(1000,ls='dashed',color='gray',alpha=0.7)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.set_ylabel('Amplitude [norm. u.]')
ax.set_xlabel('Time around Delta trough [ms]')
ax.set_xticks([0,500,1000,1500,2000])
ax.set_xticklabels(np.array([0,500,1000,1500,2000])-1000)

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
plt.close()
```

### Correlation of the average RB envelope between all pairs of arrays during EC and EO

```python
def bin_vector(vec,width=50):
    short_vec = vec[:len(vec) - len(vec)%N]
    binned_vec = short_vec.reshape(-1, N).sum(axis=1)
    return binned_vec
```

```python
N = 10

EC_corr_list = []
EO_corr_list = []
for date in all_delta_dict.keys():
    print(date)
    all_arrays = all_delta_dict[date]['All'].keys()
    all_pairs = list(combinations(all_arrays, 2))
    for pair in all_pairs:
        EC_vec_0 = all_env_dict[date]['EC'][pair[0]]
        EC_vec_1 = all_env_dict[date]['EC'][pair[1]]
        EO_vec_0 = all_env_dict[date]['EO'][pair[0]]
        EO_vec_1 = all_env_dict[date]['EO'][pair[1]]
        
        EC_0_binned = bin_vector(EC_vec_0,width=N)
        EC_1_binned = bin_vector(EC_vec_1,width=N)
        EO_0_binned = bin_vector(EO_vec_0,width=N)
        EO_1_binned = bin_vector(EO_vec_1,width=N)
        
        EC_corr_coef, _ = pearsonr(EC_0_binned,EC_1_binned)
        EO_corr_coef, _ = pearsonr(EO_0_binned,EO_1_binned)
        EC_corr_list.append(EC_corr_coef)
        EO_corr_list.append(EO_corr_coef)
```

```python
fig, ax = plt.subplots()
fig.set_figwidth(3)
fig.set_figheight(3)

ax.hist(EC_corr_list,color='darkblue',alpha=0.6,bins=21,density=True)
ax.hist(EO_corr_list,color='goldenrod',alpha=0.7,bins=21,density=True)

ax.set_xlabel('Pearson CC [norm. u.]')
ax.set_ylabel('Density [norm. u.]')

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.show()
plt.close()
```
